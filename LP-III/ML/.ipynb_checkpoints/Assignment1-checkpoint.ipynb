```python
def remove_outlier(df, col) :
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    ub = Q3 + 1.5*IQR
    lb = Q1 - 1.5*IQR
    data[col] = np.clip(data[col], lb, ub)
    return df

def remove(df, col_list) :
    for c in col_list :
        df = remove_outlier(data, c)
    return df
```

```python
data = remove(data, data.iloc[:, 0::])
```

```python
from math import *
def distance(long1, lat1, long2, lat2) :
    dist = []
    for pos in range(len(long1)):
        lo1, la1, lo2, la2 = map(radians,[long1[pos], lat1[pos], long2[pos], lat2[pos]])
        lo = lo2 - lo1
        la = la2 - la1
        a = sin(la/2)**2 + cos(la1)*cos(la2)*sin(lo/2)**2
        c = 2*asin(sqrt(a))*6371
        dist.append(c)
    return dist
```

```python
data['traveled'] = distance(data['pickup_longitude'].to_numpy(),
                            data['pickup_latitude'].to_numpy(),
                            data['dropoff_longitude'].to_numpy(),
                            data['dropoff_latitude'].to_numpy())
data.head()
```

```python
x = data.drop(['fare_amount'], axis=1)
y = data['fare_amount']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
y_pred = regression.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 10)
```

```python
rfr.fit(x_train, y_train)
Y = rfr.predict(x_test)
Y
```

```python
r2_score(Y, y_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred, y_test)
MSE = mean_squared_error(Y, y_test)
rmse = np.sqrt(mse)
RMSE = np.sqrt(MSE)
```
